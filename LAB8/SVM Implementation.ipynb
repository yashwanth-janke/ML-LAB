## Simple SVM

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC, LinearSVC, NuSVC
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
from ucimlrepo import fetch_ucirepo 
car_evaluation = fetch_ucirepo(id=19) 

X = car_evaluation.data.features 
y = car_evaluation.data.targets 
```

```python
car_evaluation.variables
```

```python
car_evaluation.variables['description'][6]
```

```python
df = pd.concat([X,y], axis=1)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
for i in X.columns:
    if X[i].dtype=='object':
        X.loc[:,i] = encoder.fit_transform(X[i])
    print(X[i].value_counts())
```

```python
X = pd.get_dummies(X, columns=X.columns, drop_first=True)
```

```python
X
```

```python
X = X.astype(int)
X
```

```python
y=df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
svm = SVC()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_svm)
print("Accuracy:", accuracy)
```

```python
lin_svm = SVC(kernel='linear')
lin_svm.fit(X_train, y_train)
y_pred_lin = lin_svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_lin)
print("Accuracy:", accuracy)
```

```python
rbf_svm = SVC(kernel='rbf')
rbf_svm.fit(X_train, y_train)
y_pred_rbf = rbf_svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_rbf)
print("Accuracy:", accuracy)
```

```python
soft_svm = SVC(kernel='linear', C=5, gamma='scale')
soft_svm.fit(X_train, y_train)
y_pred_soft = soft_svm.predict(X_test)
y_pred_soft_train = soft_svm.predict(X_train)
accuracy_on_test = accuracy_score(y_test, y_pred_soft)
accuracy_on_train = accuracy_score(y_train, y_pred_soft_train)
print("Accuracy on Test data:", accuracy_on_test)
print("Accuracy on Train data:", accuracy_on_train)
```

```python
poly_svm = SVC(kernel='poly', degree=3)
poly_svm.fit(X_train, y_train)
y_pred_poly = poly_svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_poly)
print("Accuracy:", accuracy)
```

```python
nu_svm = NuSVC(nu=0.09, kernel='rbf')
nu_svm.fit(X_train, y_train)
y_pred_nu = nu_svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_nu)
print("Accuracy:", accuracy)
```

```python
models = [svm, lin_svm, rbf_svm, soft_svm, poly_svm, nu_svm]
```

```python
for i in models:
    print(i.n_support_)
```

```python
for i in models:
    print("indices of sv for " , i)
    print(i.support_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
h = .02  
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
# title for the plots
titles = ['SVC with linear kernel',
          'LinearSVC (linear kernel)',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel']
for i, clf in enumerate(tuple(models)):
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    plt.subplot(2, 2, i + 1)
    plt.subplots_adjust(wspace=0.4, hspace=0.4)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])

plt.show()
```
